Python Pandas Tutorial 5: Handle Missing Data: fillna, dropna, interpolate

```python
df = pd.read_csv('data_frame_vaje/d_weather_data.csv', delimiter=';')
df
```

```python
## PARSE DATES:
df = pd.read_csv('data_frame_vaje/d_weather_data.csv', delimiter=';', parse_dates=['day'])
df
```

```python
## INDEX DF WITH DATE:
df = pd.read_csv('data_frame_vaje/d_weather_data.csv', delimiter=';', parse_dates=['day'])
df.set_index('day', inplace=True)
df
```

```python
## REPLACE NAN WITH VALUE:
new_df = df.fillna(0)
new_df
```

```python
## FILL VALUES WITH DICTIONARY:
new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'no event'
})
new_df
```

```python
## FILL WITH MORE REASONABLE VALUES: ffill --> forward fill
new_df = df.fillna(method='ffill')
new_df
```

```python
## FILL WITH MORE REASONABLE VALUES: bfill --> back fill
## kopira vrednosti, ki so naslednje vrednosti nazaj
new_df = df.fillna(method='bfill')
new_df
```

```python
## FILLA VREDNOSTI HORIZONTALNO (PO VRSTICI):
new_df = df.fillna(method='bfill', axis='columns')
new_df
```

```python
## FILLA VREDNOSTI FORWARD AMPAK SAMO PO 1:
new_df = df.fillna(method='ffill', limit=1)
new_df
```

```python
## FILLA VREDNOSTI Z INTERPOLACIJO:
## time ti naštima vrednosti ki so bolj bližje vrednostim,
## ki so izmerjene časovno bolj skupaj
new_df = df.interpolate(method='time')
new_df
```

```python
## DROPA VRSTICE, KI JIM MANJKAJO VSAJ ENA VREDNOST:
new_df = df.dropna()
new_df
```

```python
## DROPA VRSTICE, KI JIM MANJKAJO VSE VREDNOSTI:
new_df = df.dropna(how='all')
new_df
```

```python
## DROPA VRSTICE, KI NIMAJO NITI ENE VALID VALUES (SO VSE VREDNSOTI NAN):
new_df = df.dropna(thresh=1)
new_df
```

```python
## DROPA VRSTICE, KI NIMAJO DVEH VALID VALUES:
new_df = df.dropna(thresh=2)
new_df
```
