 Handle Missing Data: replace function

```python
## DATAFRAME WITH SPECIAL VALUES:
df = pd.read_csv('data_frame_vaje/e_weather_data.csv',delimiter=';')
```

```python
df
```

```python
## REPLACING SPEC. VALUES:
new_df = df.replace(-99999, np.NaN)
new_df
```

```python
## REPLACING VALUES:
new_df = df.replace({
    'temperature': -99999,
    'windspeed': -99999,
    'event': '0'
    },np.NaN)
new_df
```

```python
## REPLACING VALUES:
new_df = df.replace({
    -99999:np.NaN,
    '0':'Sunny',
    })
new_df
```

```python
## REGEX (odstrani definirane črke):
new_df = df.replace('[A-Za-z]','',regex=True)
new_df
```

```python
## REGEX with dictionary (odstrani definirane črke):
new_df = df.replace({
    'temperature': '[A-Za-z]',
    'event': '[A-Za-z]'
    },'',regex=True)
new_df
```

```python
## REPLACE VALUES WITH LIST OF VALUES:
new_df = df.replace(['Rain', 'Sunny'], [1, 2])
new_df
```
