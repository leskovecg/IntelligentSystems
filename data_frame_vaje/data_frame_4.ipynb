```python
## CONVERTERS:
def convert_people_cell(cell):
    if cell=='n.a.':
        return 'sam walton'
    return cell

def convert_eps_cell(cell):
    if cell=='not available':
        return None
    return cell

df = pd.read_excel('data_frame_vaje/c_stock_data.xlsx', 'c_stock_data', skiprows=1, converters={
    'people': convert_people_cell,
    'eps': convert_eps_cell    
})
df
        
```

```python
df.to_excel('data_frame_vaje/new3.xlsx','stocks') ## WRITE EXCEL
```
